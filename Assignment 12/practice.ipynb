```python
# 1) Create numpy array and perform following operation.

# Convert 1D array to 2D

import numpy as np
a = np.array([1,2,3,4,5,6])
print(a,a.shape)

a_2d = a.reshape(3,2)
print(a_2d)
```

```python
# Print Array Attributes(Like shape, dimension, data type, itemsiz)
import numpy as np
b = np.array([1,2,3,4,5,6])
print("\nShape: ",b.shape)
print("\nDimension: ",b.ndim)
print("\nDataType: ",b.dtype)
print("\nItemSize: ",b.itemsize)
```

```python
# Create a 4x4x3 array and extract value at its second set, first row and last column.

import numpy as np

arr = np.arange(1,49)
y = (arr.reshape(4,4,3))
print(y)

val = y[1,0,2]
print(val)
```

```python
# Create a 4x4 and Extract Odd Rows and Even Columns

import numpy as np
arr = np.arange(1,17).reshape(4,4)
print(arr)

res = arr[1::2,0::2]
print(res)
```

```python
# Slice the first two rows and first two columns of second set from a 4x4x3 array

import numpy as np

arr = np.arange(1,49)
y = (arr.reshape(4,4,3))
print(y)

res= y[1,0:2,0:2]
print(res)

```

```python
# Replace all odd numbers in a NumPy array with -1 by iterating using for loop [[23, 56, 78, 93],[71, 82,13, 24]]
import numpy as np

arr = np.array([[23, 56, 78, 93],
                [71, 82, 13, 24]])


rows,columns = arr.shape
for i in range(rows):
    for j in range(cols):
        if arr[i, j] % 2 != 0:   # check odd number
            arr[i, j] = -1

print(arr)



```

```python
# Get the indices of non-zero elements in an array [1, 0, 2, 8, 3, 0, 4]
import numpy as np

arr = np.array([1, 0, 2, 8, 3, 0, 41])

indices = np.nonzero(arr)

print(indices)
```

```python
# Perform arithmetic operations on two NumPy arrays element-wise

# Add two NumPy arrays element by element. Multiply two NumPy arrays element by element.

import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])

add = arr1+arr2
print(add)
mul = arr1*arr2
print(mul)
```

```python
# Write a code to compute the dot product of two NumPy arrays arr1 [15, 20, 25] arr2 [10,40,37]

import numpy as np

arr1 = np.array([15, 20, 25])
arr2 = np.array([10, 40, 37])

res = np.dot(arr1,arr2)
print("Dot Product = ",res)
```
