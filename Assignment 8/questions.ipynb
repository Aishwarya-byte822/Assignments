```python
#  q1 Convert a series of date-strings to a timeseries?

import pandas as pd
df = pd.DataFrame({
    'Col1':range(12),
    'Col2':['A','A','A','B','B','B','C','C','C','D','D','D'],
    'date':pd.to_datetime(['2026-05-21','2026-05-22','2026-05-23']*4)
    })
print(df)
pivoted = df.pivot(
    index= 'date',
    columns= 'Col2',
    values='Col1'
)
print("Pivoted DataFrame: \n",pivoted)
```

```python
#q2 Create two DataFrames, df1 and df2, with a common column (e.g., 'ID').
import pandas as pd
df1 = pd.DataFrame({
    'Name': ['A1','A2','A3','A4','A5'],
    'Subject': ['sub1','sub2','sub4','sub5','sub6'],
    'ID': [1,2,3,4,5]
})

df2 = pd.DataFrame({
    'Name':['B1','B2','B3','B4','B5'],
     'Subject': ['sub2','sub4','sub3','sub6','sub5'],
     'ID':[3,4,5,6,7]
})
# print(df1)
# print(df2)

# Perform an inner merge on this common column and display the resulting DataFrame.
res = df1.merge(df2,on='ID',how = 'inner')
print(res)

# Perform a left join of df1 and df2 on the 'ID' column. Explain how missing values are handled in the resulting DataFrame. Right Join and Index-Based Join.

res1 = df1.merge(df2, on='ID',how = 'left') 
print(res1)
res2 = df1.merge(df2,on = 'ID', how = 'right')
print(res2)


df1_indexed = df1.set_index('ID')
df2_indexed = df2.set_index('ID')


result = df1_indexed.join(df2_indexed, lsuffix='_left', rsuffix='_right')
print(result)

```

```python
# Perform a right join using pd.merge() on a common column, then perform a join using df.join() based on the index. Compare the results. Merging with Multiple Keys.

import pandas as pd
df1 = pd.DataFrame({
    'Name': ['A1','A2','A3','A4','A5'],
    'Subject': ['sub1','sub2','sub4','sub5','sub6'],
    'ID': [1,2,3,4,5]
})

df2 = pd.DataFrame({
    'Name': ['B1','B2','B3','B4','B5'],
    'Subject': ['sub2','sub4','sub3','sub6','sub5'],
    'ID': [3,4,5,6,7]
})

res = df1.merge(df2, on='ID', how = 'inner')
print(res)

df1_indexed = df1.set_index('ID')
df2_indexed = df2.set_index('ID')

index_res = df1_indexed.join(df2_indexed, how='inner', lsuffix='df1', rsuffix='df2')
print(index_res)

multi = df1.merge(df2,on = ['ID','Subject'],how = 'inner')
print(multi)
```

```python
# Create three DataFrames. Vertically concatenate two of them using pd.concat(), then merge the resulting DataFrame with the third DataFrame on a common key. T Understand join() vs. merge().

import pandas as pd

df1 = pd.DataFrame({
    'Name': ['A1','A2','A3','A4','A5'],
    'Subject': ['sub1','sub4','sub5','sub2','sub3'],
    'ID':[1,2,3,4,5]
})

df2 = pd.DataFrame({
    'Name':['B1','B2','B3','B4','B5'],
    'Subject':['sub4','sub2','sub3','sub1','sub5'],
    'ID':[3,4,5,6,7]
})

df3 = pd.DataFrame({
    'Name':['C1','C2','C3','C4','C5'],
    'Subject':['sub6','sub2','sub5','sub1','sub2'],
    'ID':[3,4,2,6,5]
})

result = pd.concat([df1,df2], ignore_index = True)
print(result)

res = df3.merge(result, on='ID', how = 'inner')
print(res)

```
